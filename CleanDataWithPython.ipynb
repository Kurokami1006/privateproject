### Clean Course_id column

```python
# Check
df['course_id'].describe()
```

```python
# Tìm ra giá trị trong cột có ký tự là chữ
df[df['course_id'].str.match(r'.*[a-zA-Z].*') == True]
```

```python
# replace hết đóng trên để quy chuẩn dạng sang số hết - Cột này vẫn là Object
df.loc[df["course_id"] == "1271182A", "course_id"] = "1271182"
df.loc[df["course_id"] == "1157298B", "course_id"] = "1157298"
df.loc[df["course_id"] == "715476C", "course_id"] = "715476"
```

```python
# ép kiểu object sang int64
#df.astype({'course_id': 'int64'}).dtypes   # Cách 1
df["course_id"].astype(str).astype(int)     # Cách 2
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)
### Clean Course_title column

```python
# Lọc ra các giá trị có xuống dòng trong cột 'course_title'
df[df['course_title'].str.contains(r'\n') == True]
```

```python
# Loại bỏ tất cả các giá trị xuống dòng
df = df.replace(r'\n',' ', regex=True)
```

```python
# Recheck
df[df['course_title'].str.contains(r'\n') == True]
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)
### Clean Subject column

```python
# Lọc ra các dự liệu bị wrong data
df.groupby('subject')['subject'].count()
```

```python
# Replace wrong data
df.loc[df["subject"] == "Business Ffinance", "subject"] = "Business Finance"
df.loc[df["subject"] == "Graphic  Design", "subject"] = "Graphic Design"
df.loc[df["subject"] == "Musical Instrument", "subject"] = "Musical Instruments"
```

```python
# Recheck
df.groupby('subject')['subject'].count()
```
